# Research Question

How do demographic factors such as race-based population density, median income, and neighborhood mortality rates influence EMS response times across different communities within San Francisco?

## Background and Prior Work

Emergency Medical Services (EMS) are an important part of the healthcare system, providing immediate medical attention and transportation to care for patients experiencing medical events. Response times are an important faucet of this, with quick emergency response being key to save lives. This project examines potential relationships between race-based population density, median income, neighborhood mortality rates, and EMS response times across San Francisco communities.

In urban environments like San Francisco, multiple factors can influence EMS response times: traffic patterns, geographic features, dispatch protocols, ambulance station locations, and resource allocation. However, factors such as socioeconomic status, race, ethnicity, education level, and neighborhood characteristics can also effect health outcomes. 

Previous studies have examined patterns in EMS response times. Hsia et al. observed response times approximately 10 percent longer in lower-income neighborhoods compared to wealthier areas, even after controlling for factors like time of day.<a name="cite_ref-1" id="cite_ref-1"></a><a href="#cite_note-1"><sup>[1]</sup></a> In cases of cardiac arrest, these differences may affect survival rates for patients from various income levels.<a name="cite_ref-2" id="cite_ref-2"></a><a href="#cite_note-2"><sup>[2]</sup></a> The CDC has documented geographic and income-based variations in emergency medical services across the country.<a name="cite_ref-3" id="cite_ref-3"></a><a href="#cite_note-3"><sup>[3]</sup></a>

This disparity in emergency response appears to follow broader patterns of health inequality in San Francisco, where health indicators correlate strongly with demographic factors.<a name="cite_ref-4" id="cite_ref-4"></a><a href="#cite_note-4"><sup>[4]</sup></a> A 2019 assessment identified health challenges concentrated in an "upside down T" pattern corresponding to neighborhoods with lower-income populations.<a name="cite_ref-4b" id="cite_ref-4b"></a><a href="#cite_note-4"><sup>[4]</sup></a> These same areas experience disproportionately higher rates of cardiovascular disease requiring emergency intervention, particularly in southeast San Francisco.<a name="cite_ref-5" id="cite_ref-5"></a><a href="#cite_note-5"><sup>[5]</sup></a> 

The higher cardiovascular disease burden among Black/African American residents (with hospitalization rates 3-5 times higher than other demographic groups)<a name="cite_ref-5b" id="cite_ref-5b"></a><a href="#cite_note-5"><sup>[5]</sup></a> suggests these communities may have greater need for rapid emergency services, yet preliminary evidence indicates they may face longer wait times. This contradiction between need and service quality could increase existing health disparities and contribute to the differing mortality rates observed in these neighborhoods.

These neighborhood-level disparities likely stem from historical policy decisions such as redlining, which systematically affected community investment patterns, including healthcare infrastructure development.<a name="cite_ref-6" id="cite_ref-6"></a><a href="#cite_note-6"><sup>[6]</sup></a> Research has identified strong correlations between current mortality patterns and areas historically subjected to discriminatory policy treatment,<a name="cite_ref-7" id="cite_ref-7"></a><a href="#cite_note-7"><sup>[7]</sup></a> suggesting that today's EMS response inequalities may reflect decades of disadvantage in affected communities.

Our aim is to quantify the relationships between these socioeconomic factors and EMS response times in an effort to better understand how EMS response times are impacted by demographics, and the ways they potentially impact disadvantaged communities.

# Hypothesis


We hypothesize that areas with the greatest population density, highest percentage of BIPOC residents (specifically Black populations), and lower median income will demonstrate significantly longer EMS response times in San Francisco. Specifically, we predict a strong positive correlation between response times and percentage of BIPOC residents, and a strong negative correlation between response times and median income. we hypothesize these demographic variables to explain substantially more variance in response times (R² > 0.6) than potential confounding variables such as traffic patterns or weather.

# Ethics & Privacy

Our project examines potential disparities in emergency medical services across San Francisco neighborhoods, which inherently involves complex ethical considerations:

While we're using publicly available datasets from San Francisco government sources, we recognize several ethical concerns:

- The EMS calls dataset may contain sensitive incident locations that, even if anonymized, could potentially be re-identified when combined with other information
- We commit to using this data only in aggregate form by neighborhood rather than by specific addresses or incidents
- We'll adhere strictly to the terms of use of all datasets and avoid any attempts to re-identify individuals

Furthermore, several biases could affect our analysis:

1. **Selection bias**: EMS call data only includes incidents where EMS was actually called, potentially missing communities where residents may avoid calling emergency services due to distrust, cultural barriers, or cost concerns
2. **Reporting bias**: Response time measurement may not be standardized across all neighborhoods or may be recorded differently depending on the responders

All in all, we acknowledge that our analysis could reveal sensitive disparities, and we commit to conducting this research with the goal of contributing to a better understanding of healthcare access rather than simply documenting inequities without context.
