# Research Question

How do demographic factors such as population count by race and median income correlate with Emergency Medical Services (EMS) response times (measured from ambulance's dispatch to it's arrival at the hospital) across different communities within San Francisco, and to what extent do such factors predict disparities in response times?

# Background and Prior Work

Emergency Medical Services (EMS) are an important part of the healthcare system, providing immediate medical attention and transportation to care for patients experiencing medical events. Response times are an important faucet of this, with quick emergency response being key to save lives. This project examines potential relationships between race-based population density, median income, neighborhood mortality rates, and EMS response times across San Francisco communities.

In urban environments like San Francisco, multiple factors can influence EMS response times: traffic patterns, geographic features, dispatch protocols, ambulance station locations, and resource allocation. However, factors such as socioeconomic status, race, ethnicity, education level, and neighborhood characteristics can also effect health outcomes.

Previous studies have examined patterns in EMS response times. Hsia et al. observed response times approximately 10 percent longer in lower-income neighborhoods compared to wealthier areas, even after controlling for factors like time of day.<a name="cite_ref-1" id="cite_ref-1"></a><a href="#cite_note-1"><sup>[1]</sup></a> In cases of cardiac arrest, these differences may affect survival rates for patients from various income levels.<a name="cite_ref-2" id="cite_ref-2"></a><a href="#cite_note-2"><sup>[2]</sup></a> The CDC has documented geographic and income-based variations in emergency medical services across the country.<a name="cite_ref-3" id="cite_ref-3"></a><a href="#cite_note-3"><sup>[3]</sup></a>

This disparity in emergency response appears to follow broader patterns of health inequality in San Francisco, where health indicators correlate strongly with demographic factors.<a name="cite_ref-4" id="cite_ref-4"></a><a href="#cite_note-4"><sup>[4]</sup></a> A 2019 assessment identified health challenges concentrated in an "upside down T" pattern corresponding to neighborhoods with lower-income populations.<a name="cite_ref-4b" id="cite_ref-4b"></a><a href="#cite_note-4"><sup>[4]</sup></a> These same areas experience disproportionately higher rates of cardiovascular disease requiring emergency intervention, particularly in southeast San Francisco.<a name="cite_ref-5" id="cite_ref-5"></a><a href="#cite_note-5"><sup>[5]</sup></a>

The higher cardiovascular disease burden among Black/African American residents (with hospitalization rates 3-5 times higher than other demographic groups)<a name="cite_ref-5b" id="cite_ref-5b"></a><a href="#cite_note-5"><sup>[5]</sup></a> suggests these communities may have greater need for rapid emergency services, yet preliminary evidence indicates they may face longer wait times. This contradiction between need and service quality could increase existing health disparities and contribute to the differing mortality rates observed in these neighborhoods.

These neighborhood-level disparities likely stem from historical policy decisions such as redlining, which systematically affected community investment patterns, including healthcare infrastructure development.<a name="cite_ref-6" id="cite_ref-6"></a><a href="#cite_note-6"><sup>[6]</sup></a> Research has identified strong correlations between current mortality patterns and areas historically subjected to discriminatory policy treatment,<a name="cite_ref-7" id="cite_ref-7"></a><a href="#cite_note-7"><sup>[7]</sup></a> suggesting that today's EMS response inequalities may reflect decades of disadvantage in affected communities.

Our aim is to quantify the relationships between these socioeconomic factors and EMS response times in an effort to better understand how EMS response times are impacted by demographics, and the ways they potentially impact disadvantaged communities. By quantifying this relationship, we can take steps in concretely identifying the issues in response time, using that result to aid us in figuring out future directions of research in this area. Furthermore, our analysis will examine a 4 year period in San Francisco, allowing us to determine whether or not the inequalities in EMS response times have improved or worsened. By including the COVID-19 pandemic period, our study uniquely examines how emergency systems respond to extreme stress and whether existing neighborhood disparities were exacerbated during this public health crisis.

# Hypothesis


We hypothesize that in San Francisco, areas with a higher percentage of Black, Indigenous, and People of Color (BIPOC) residents—particularly Black populations—and lower median incomes will experience longer Emergency Medical Services (EMS) response times (measured from dispatch to hospital arrival). These disparities may be attributed to historical redlining and disinvestment, which could have resulted in under-resourced infrastructure, leading to fewer emergency service stations in predominantly BIPOC and low-income neighborhoods. As such, we predict a positive correlation between response times and BIPOC population percentage, and a negative correlation between response times and median income. We can use the census.gov API to retrieve median income data for specific zipcodes (neighborhoods in San Francisco). This data can be joined back to our EMS response dataset, as that data includes zip codes as a column

Request the census.gov API for data about median income for each of the above zip codes, from 2020-2023

```python
# creating: function to hep us get the median income
def get_median_income(zip_code, year):
    """
    gets median household income for a zip code and year
    """
    base_url = f"https://api.census.gov/data/{year}/acs/acs5"
    params = {
        "get": "B19013_001E", # median household income code
        "for": f"zip code tabulation area:{zip_code}"
    }

    full_url = f"{base_url}?{urlencode(params)}"
    print(f"requesting: {full_url}") # uncomment for url output

    try:
        response = requests.get(base_url, params=params, timeout=15)
        response.raise_for_status() # raise http errors
        data = response.json()

        # extract: income from response data
        if len(data) > 1 and isinstance(data[1], list) and len(data[1]) > 0:
            income_value = data[1][0]
            return int(income_value) if income_value not in (None, "null", "") else None
        else:
             print(f"no data found for zip {zip_code}, year {year}. data: {data}")
             return None

    except requests.exceptions.Timeout:
        print(f"timeout for zip {zip_code}, year {year}. url: {full_url}")
        return None
    except requests.exceptions.HTTPError as http_err:
        print(f"http error for zip {zip_code}, year {year}: {http_err} - url: {full_url}")
        return None
    except requests.exceptions.JSONDecodeError as json_err:
        print(f"json error for zip {zip_code}, year {year}: {json_err} - content: {response.text[:200] if response else 'no response'} - url: {full_url}")
        return None
    except Exception as e:
        print(f"error for zip {zip_code}, year {year}: {e} - url: {full_url}")
        return None
    finally:
        # pause briefly after each request
        time.sleep(0.2)
```

```python
# DO NOT RUN THE CODE BELOW ! (THIS CODE REQUESTS DATA FROM THE CENSUS.GOV API TAKING 10 MIN+)

RUN = False

if RUN:
    raw_unique_zips_array = ems_responses['zipcode'].unique()
    print(f"Raw unique zip codes from dataframe: {raw_unique_zips_array}")

    # standardize the unique zip codes
    cleaned_unique_zips = []
    for item in raw_unique_zips_array:
        if pd.isna(item): # skip missing values (none/nan)
            continue
        zip_str = str(item)
        if zip_str.endswith('.0'):
            zip_str = zip_str[:-2] 
        cleaned_unique_zips.append(zip_str)

    test_zips = cleaned_unique_zips # use cleaned zips for API calls

    print(f"Cleaned and standardized {len(test_zips)} unique zip codes.")

    test_years = [2020, 2021, 2022, 2023] # years for data

    # use threaded requests for speed
    results = []
    tasks_to_submit = []

    start_time = time.time()

    # prepare zip-year task list
    for zip_code in test_zips:
        for year in test_years:
            tasks_to_submit.append((zip_code, year))

    max_workers = 5 # limit concurrent requests to census api
    print(f"\nStarting data fetch for {len(tasks_to_submit)} tasks using {max_workers} workers...")
    print("Waiting for results...")

    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        # submit tasks and map futures to their params
        future_to_params = {
            executor.submit(get_median_income, zip_val, year_val): (zip_val, year_val)
            for zip_val, year_val in tasks_to_submit
        }

        # process results as they complete
        for future in concurrent.futures.as_completed(future_to_params):
            zip_code, year = future_to_params[future]
            try:
                income = future.result() # get result or exception
                results.append({
                    'zipcode': zip_code,
                    'year': year,
                    'median_income': income
                })
            except Exception as exc:
                print(f"Task for zip {zip_code}, year {year} failed: {exc}")
                results.append({
                    'zipcode': zip_code,
                    'year': year,
                    'median_income': None
                })

    end_time = time.time()
    print(f"\nFinished processing {len(results)} results.")
    print(f"Total time taken: {end_time - start_time:.2f} seconds")

    # results df
    income_df = pd.DataFrame(results)
    income_df = income_df.sort_values(by=['zipcode', 'year']).reset_index(drop=True)

    print("\n--- Results Dataframe ---")
    print(income_df)

    # save results to csv
    output_filename = 'median_income_by_zip_year.csv'
    try:
        income_df.to_csv(output_filename, index=False)
        print(f"\nresults saved to {output_filename}")
    except Exception as e:
        print(f"\nError saving results to csv: {e}")
```

```python
# identify: any neighborhoods that don't match exactly between datasets
race_neighborhoods = set(neighborhood_race_df['neighborhood'].unique())
ems_neighborhoods = set([n for n in ems_responses['neighborhood'].unique() if isinstance(n, str)])  # exclude nan

# check: for exact matches and mismatches
exact_matches = race_neighborhoods.intersection(ems_neighborhoods)
in_race_not_ems = race_neighborhoods - ems_neighborhoods
in_ems_not_race = ems_neighborhoods - race_neighborhoods

# check: this allows us to see the information written out clearly
print(f"Exact matches: {len(exact_matches)} neighborhoods")
print(f"In race data but not in EMS data: {len(in_race_not_ems)} neighborhoods: {in_race_not_ems}")
print(f"In EMS data but not in race data: {len(in_ems_not_race)} neighborhoods: {in_ems_not_race}")
```

```python
# create: a simple mapping for the one non-matching neighborhood
crosswalk = {
    'The Farallones': None  # this neighborhood doesn't exist in EMS data, so we will drop it from our analysis (only 8 rows)
}

# create: standardized neighborhood columns in both dataframes
# for the race data
neighborhood_race_df['standardized_neighborhood'] = neighborhood_race_df['neighborhood'].apply(
    lambda x: crosswalk.get(x, x)  # use mapping if available, otherwise keep as is
)

# add: standardized column to EMS data (all neighborhoods already match)
ems_responses['standardized_neighborhood'] = ems_responses['neighborhood']

# create: neighborhood summary from race data
# this aggregates demographic data by neighborhood
neighborhood_summary = pd.DataFrame()

# get total population by neighborhood
total_pop_by_neighborhood = neighborhood_race_df.groupby('standardized_neighborhood')['population'].sum().reset_index()
neighborhood_summary['standardized_neighborhood'] = total_pop_by_neighborhood['standardized_neighborhood']
neighborhood_summary['total_population'] = total_pop_by_neighborhood['population']

# create: racial percentage columns - create one column per race category
for race_category in race_categories:
    # get a short name for the column
    short_name = race_category.split(',')[0].strip().lower().replace(' ', '_')
    
    # filter: data for this race
    race_data = neighborhood_race_df[neighborhood_race_df['race'] == race_category]
    race_data = race_data[['standardized_neighborhood', 'population']]
    
    # rename: column for the merge
    race_data = race_data.rename(columns={'population': f'pop_{short_name}'})
    
    # merge: with summary
    neighborhood_summary = neighborhood_summary.merge(
        race_data, 
        on='standardized_neighborhood', 
        how='left'
    )
    
    # calculate: percentage
    neighborhood_summary[f'pct_{short_name}'] = (
        neighborhood_summary[f'pop_{short_name}'] / 
        neighborhood_summary['total_population'] * 100
    ).fillna(0)

# create: minority percentage (everyone except white alone)
white_col = [col for col in neighborhood_summary.columns if 'pop_white_alone' in col]
if white_col:
    white_col = white_col[0]
    neighborhood_summary['minority_population'] = neighborhood_summary['total_population'] - neighborhood_summary[white_col]
    neighborhood_summary['pct_minority'] = (
        neighborhood_summary['minority_population'] / 
        neighborhood_summary['total_population'] * 100
    ).fillna(0)

# join: with EMS data
merged_data = ems_responses.merge(
    neighborhood_summary,
    on='standardized_neighborhood',
    how='left'
)

# check the join results
print(f"Total EMS responses: {len(ems_responses)}")
print(f"Merged data rows: {len(merged_data)}")
print(f"Missing demographic data: {merged_data['total_population'].isna().sum()} rows")

# preview the merged data
print("\nMerged data preview:")
print(merged_data.head())
```

# Ethics & Privacy

Our project examines potential disparities in emergency medical services across San Francisco neighborhoods, which inherently involves complex ethical considerations.

While we're using publicly available datasets from San Francisco government sources, we recognize several ethical concerns:
- The EMS calls dataset may contain sensitive incident locations that, even if anonymized, could potentially be re-identified when combined with other information
- We'll adhere strictly to the terms of use of all datasets and avoid any attempts to re-identify individuals

Furthermore, several biases could affect our analysis:


**Selection bias:** EMS call data only includes incidents where EMS was actually called, potentially missing communities where residents may avoid calling emergency services due to distrust, cultural barriers, or cost concerns. Additionally, the number of EMS calls may vary greatly per neighborhood depending on the population, because a neighborhood with high population is likely to experience more EMS calls. We can address this by using multivariate regression models that control for call volume and population differences when analyzing response times, rather than equalizing neighborhood representation which could distort important patterns. This approach allows us to identify disparities while accounting for legitimate operational differences across neighborhoods.

**Pandemic-related bias:** The COVID-19 pandemic created disruptions to EMS systems that varied by neighborhood based on infection rates, hospital capacity, and staffing challenges. To address this temporal bias, we will analyze pre-pandemic, acute crisis, and recovery phases separately, and documenting any known operational changes in EMS protocols during high-surge periods. This approach will help distinguish pandemic-specific effects from underlying neighborhood disparities.

The selected timeframe (2020-2023) was chosen in order to capture three distinct phases:
1) The immediate pre-pandemic economy (beginning 2020)
2) The acute crisis period (2020-2021)
3) The post-vaccine recovery (2022-2023)
This allows us to isolate pandemic-specific effects from longer-term economic trends

For Dataset #3 (Census data), we selected the most recent available dataset (2019–2023) as it provides the closest proximity to our study period (2020–2023). While this dataset includes 2019—a pre-pandemic year—it remains preferable to older alternatives (e.g., 2014–2019), which would introduce greater bias. The 2019 data cannot be isolated due to aggregation in the source, but we anticipate minimal impact on neighborhood population analyses.

A further consideration regarding the median income data from the Census API is its granularity over time. The API provides 5-year estimates for ZCTAs, not precise annual figures for each year (2020-2023). Each year's estimate (e.g., the 2020 estimate is based on data collected between 2016 and 2020) is a rolling average influenced by data from both before and during our study period, and even including pre-pandemic years for the earlier estimates (like the 2020 estimate). This inherent smoothing effect will likely dampen or mask any sharp year-to-year income fluctuations that may have occurred specifically due to the COVID-19 pandemic. While this data is the most relevant available at the ZCTA level for our timeframe, it means our analysis may be better suited to examining the correlation between average income levels over the study period and EMS response times, rather than precisely isolating the impact of pandemic-induced *changes* in income on response times within a given year. We will discuss this data characteristic and its implications for interpreting year-over-year trends in our final analysis.

**Reporting bias:** Response time measurements may vary across neighborhoods due to inconsistent recording practices among EMS teams. To address this, we can examine official EMS documentation outlining standardized protocols (including when dispatchers initiate timing, whether systems use automated timestamps, and how times are recorded at hospitals). By documenting these official standards and identifying any procedural variations, we'll be able to account for measurement discrepancies that could affect neighborhood comparisons in our analysis.

Some unintended consequences of our study include stigmatizing communities where EMS response times are high, creating negative perceptions around these communities. Furthermore, findings might lead to a misallocation of resources if resources are moved from other underserved communities to improve overall response time averages. All in all, we acknowledge that our analysis could reveal sensitive disparities, and we commit to conducting this research with the goal of contributing to a better understanding of healthcare access rather than simply documenting inequities without context.
